```python
csv_file = open("booking.csv", 'w')
```

```python
csv_writer = csv.writer(csv_file)
csv_writer.writerow(['properties', 'address', 'distance', 'price', 'reviews','ratings', 'description', 'photos'])
```

```python
properties = soup.find_all("a",{"class":"bui-card__header_full_link_wrap"})

address = soup.find_all("p", {"class":"bui_card_header_full_link_wrap"})

distance = soup.find_all("span", {"data-testid":"property-card"})

prices = soup.find_all("div", {"class":"bui-price-display__value"})

reviews = soup.find_all("div", {"class":"bui-review-score__text"})

ratings = soup.find_all("span",{ "class":"review-score-badge"})

description = soup.find_all("p", {"class": "bui-card__text hotel-card__text--wrapped bui-spacer--small"})

photos = soup.find_all("img")
```

```python
for h in properties:
    print(h.text)

for a in address:
    print(a['span'])

for b in distance:
    print(b.text)

for p in prices:
    print(p.text)

for t in reviews:
    print(t.text)

for r in ratings:
    print(r.text)

for c in description:
    print(c.text)

for s in photos:
    print(s['src'])

    print()

    csv_writer.writerow([properties, address, distance, prices, reviews, ratings, description, photos])
csv_file.close();
```
